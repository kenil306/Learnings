StockistId where hqId is missing

```python
def null(df):

    # Filter rows where any of the key location IDs are missing
    missing = df[df[['hqId']].isna().any(axis=1)]

    # Get unique combinations of stockistId and chemistId with missing info
    result = missing[['stockistId', 'hqId']].drop_duplicates().reset_index(drop=True)

    return result
new_df = null(df)
print(new_df.to_string())
```

```python
X = cd.data_df.drop('amount', axis=1)
y = cd.data_df['amount']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
# X_train['year'] = pd.to_datetime(X_train['year'], errors='coerce')

X_train['invoice_year'] = X_train['invoiceDate'].dt.year
X_train['invoice_month'] = X_train['invoiceDate'].dt.month
X_train['invoice_day'] = X_train['invoiceDate'].dt.day
X_train['invoice_weekday'] = X_train['invoiceDate'].dt.weekday

# Then drop the original datetime column
X_train = X_train.drop(columns=['invoiceDate'])
```

```python
# X_train['week_of_year'] = pd.to_datetime(X_train['week_of_year'], errors='coerce')

X_test['invoice_year'] = X_test['invoiceDate'].dt.year
X_test['invoice_month'] = X_test['invoiceDate'].dt.month
X_test['invoice_day'] = X_test['invoiceDate'].dt.day
X_test['invoice_weekday'] = X_test['invoiceDate'].dt.weekday

X_test = X_test.drop(columns=['invoiceDate'])
```

```python
X_train.info()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=10, max_depth=12)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("MSE:", mean_squared_error(y_test, y_pred))               # MSE tells you how far off your predictions are; lower the better
print("RÂ²:", r2_score(y_test, y_pred))                          # RÂ² tells you how useful the model is

###
r2 = r2_score(y_test, y_pred)
accuracy_percentage = r2 * 100
print("Model Accuracy (%):", accuracy_percentage)
```
